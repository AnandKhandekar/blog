Descriptive statistics involves summarizing and organizing data. The data analyzed by descriptive statistics is usually a sample of a population of data. Measures such as the mean, median and mode describe the data; the data can also be summarized visually with graphs such as histograms. scatter plots, bar charts, polar (Phaser ) ....

* ## Central Tendencies
A measure of central tendency is a summary measure that attempts to describe a whole set of data with a single value that represents the middle or center of its distribution. Measures of central tendency estimate the typical value in a data set.

Source[Measures of Central Tendency](https://www.abs.gov.au/websitedbs/D3310114.nsf/Home/Statistical+Language+-+measures+of+central+tendency)  from the Australian Bureau of Statistics

* ## Scenario
You are looking to purchase a home in a certain neighborhood. You are given a data set of 91 houses that were sold over the past year in and around the neighborhood. Assuming that the houses are relatively similar, you want to purchase a house for the best price in the neighborhood.

*  ## Arithmetic Mean (Average)
The arithmetic mean (average) is a measure of central tendency of a data set.

If the data set is symmetrically distributed (approximating a bell shape when the data is plotted on a histogram), then the mean can be a reliable tool to understand the central tendency of the data set.

However the mean can be skewed by outliers if the data is not symmetrically distributed. For example if a few houses sold significantly more than the other houses in the neighborhood the average would be skewed to the more expensive side of the distribution.

\begin{align}
Arithmetic \cdot Mean = \frac{1}{n} \cdot \sum_{i=1}^n x_i
\end{align}

```python
#collapse-show
# custom function to calculate MEAN
def mean_calc(data_set):
    return sum(data_set)/len(data_set)
```

```python
#use Numpy to check if our function works propoerly
print(mean_calc(house_prices) == np.mean(house_prices))
```

```python
# store the calculated MEAN in a variable
mean_house_price = mean_calc(house_prices)
```

```python
#collapse-show
# plot the histogram
plt.hist(house_prices,edgecolor='black',color='purple')
plt.axvline(mean_house_price,color='red',lw=3,label=f'Average (Mean) House Price: ${mean_house_price:,.0f}')
plt.xlabel('House Price')
plt.xticks([num for num in range(100000,450000,50000)],
          ['$'+str(num)[:3] + ',' +str(num)[3:] for num in range(100000,450000,50000)])
plt.ylabel('Count')
plt.title('House Prices (Positive Skew)')
plt.legend();
```

* ## Median
The median is the middle number (if the number of data points is odd). It is found by ordering all data points and picking out the one in the middle. If there are two middle numbers (the number of data points is even), the median is calculated by taking the mean of those two numbers.

If the data set is skewed (non-symmetric), the median is a better indicator of central tendency relative to the mean.

\begin{equation}
  \Large{\text{Median}} =
    \begin{cases}
      \Large{y_{[(n+1)/2]}} & \text{if $n$ is odd}\\
      \Large{\frac{y_{(n/2)} + y_{(n/2+1)}}{2}} & \text{if $n$ is even}
    \end{cases}       
\end{equation}

Source: Mean, median, and mode review from [Khan Academy](https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/mean-median-basics/a/mean-median-and-mode-review)

```python
#collapse-show
# write a custom function to calculate the Median
def median_calc(data_set):
    sorted_data_set = list(sorted(data_set))  # not used np.sort()
    data_set_length = len(sorted_data_set)
    
    if data_set_length == 1:
        median = data_set[0]
    elif data_set_length % 2 != 0:
        median = sorted_data_set[int((data_set_length)/2)]
    else:
        median = (sorted_data_set[int(data_set_length/2)-1]+sorted_data_set[int(data_set_length/2)])/2
    
    return median
```

```python
# use Numpy to verify if our custome function is correct
median_calc(house_prices) == np.median(house_prices)
median_house_price = median_calc(house_prices)
```

```python
#collapse-show
plt.hist(house_prices,edgecolor='black',color='purple')
plt.axvline(mean_house_price,color='red',lw=3,label=f'Average (Mean) House Price: ${mean_house_price:,.0f}')
plt.axvline(median_house_price,color='orange',lw=3,label=f'Median House Price: ${median_house_price:,.0f}')
plt.xticks([num for num in range(100000,450000,50000)],
          ['$'+str(num)[:3] + ',' +str(num)[3:] for num in range(100000,450000,50000)])
plt.xlabel('House Prices')
plt.ylabel('Count')
plt.title('House Prices (Positive Skew)')
plt.legend();
```

* ## Mode
The mode is the most commonly occurring value in a distribution. The mode can only be calculated for a data set that contains discrete values.

```python
# custom function to find the MODE
def mode_calc(data_set):
    return max(set(data_set), key=data_set.count)

mode_house_price = mode_calc(house_prices)
```

```python
# check if our custom function is accurate
mode_house_price == statistics.mode(house_prices)
```

```python
height_unique = list(set(house_prices))
height_count = [house_prices.count(height) for height in height_unique]
```

```python
fig = px.scatter(x=height_unique, y=height_count,size=height_count,
                 title='House Prices',
                 labels={'x':'House Prices','y':'Count'})

fig.update_yaxes(tickmode = 'array',
                 tickvals = np.arange(0,5))

fig.show()
```

* ## Describing the data
A symmetric distribution is one in which the 2 halves of the histogram appear as mirror-images of one another. A skewed non-symmetric distribution is a distribution in which there is no such mirror-imaging.

For skewed distributions, it is quite common to have one tail of the distribution considerably longer or drawn out relative to the other tail. A skewed right distribution is one in which the tail is on the right side. A skewed left distribution is one in which the tail is on the left side. The above histogram is for a distribution that is skewed right.

From our data set, we can see the house prices are skewed right (Mode < Median < Mean), telling us that the prices of the houses are skewing to the more expensive part of the range.

```python
#collapse-show
plt.hist(house_prices,edgecolor='black',color='purple')
plt.axvline(mean_house_price,color='red',lw=3,label=f'Average (Mean) House Price: ${mean_house_price:,.0f}')
plt.axvline(median_house_price,color='orange',lw=3,label=f'Median House Price: ${median_house_price:,.0f}')
plt.axvline(mode_house_price,color='pink',lw=3,label=f'Mode House Price: ${mode_house_price:,.0f}')
plt.xticks([num for num in range(100000,450000,50000)],
          ['$'+str(num)[:3] + ',' +str(num)[3:] for num in range(100000,450000,50000)])
plt.xlabel('House Price')
plt.ylabel('Count')
plt.title('House Prices (Positive Skew)')
plt.legend();
```
